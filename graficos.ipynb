```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o conjunto de dados e armazenar em um DataFrame
df = pd.read_csv('android-games.csv')

# Selecionar os dados de interesse (categoria e nmero de estrelas)
df = df[['category', '5 star ratings']]

# Agrupar os dados por categoria e calcular o total de estrelaas
df = df.groupby(['category']).sum()

# Criar um grfico de pizza
fig, ax = plt.subplots()

# Adicionar os dados ao grfico de pizza
ax.pie(df['5 star ratings'], labels=df.index, autopct='%1.1f%%')

# Adicionar ttulo ao grfico
plt.title('Porcentagem de 5 stars por categoria de jogo na Google Play Store')

# Mostrar o grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o conjunto de dados e armazenar em um DataFrame
df = pd.read_csv('android-games.csv')

# Selecionar os dados de interesse (ttulo e nmero total de avaliaes)
df = df[['title', 'total ratings']]

# Ordenar os dados por nmero total de avaliaes
df = df.sort_values(by=['total ratings'], ascending=False)

# Limitar o nmero de jogos a serem exibidos
num_jogos = 10
df = df.head(num_jogos)

# Criar um grfico de linhas
fig, ax = plt.subplots()

# Adicionar os dados ao grfico de linhas
ax.plot(df['title'], df['total ratings'])

# Adicionar rtulos ao eixo x e ao eixo y
plt.xlabel('Ttulo do jogo')
plt.ylabel('Total de avaliaes')

# Girar os rtulos do eixo x em 90 graus
plt.xticks(rotation=90)

# Adicionar ttulo ao grfico
plt.title(f'{num_jogos} jogos mais avaliados na Google Play Store')

# Mostrar o grfico
plt.show()
```

```python
import squarify
import pandas as pd

df = pd.read_csv('android-games.csv')
# Agrupa os jogos por categoria
grouped = df.groupby(['category']).size().reset_index(name='counts')

# Ordena as categorias pela contagem
grouped = grouped.sort_values('counts', ascending=False)

# Cria uma lista com as contagens de cada categoria
counts = grouped['counts'].tolist()

# Cria uma lista com os nomes de cada categoria
labels = grouped['category'].tolist()

# Calcula o tamanho de cada retngulo no grfico de rvore
sizes = [(count / sum(counts)) * 100 for count in counts]

# Cria o grfico de rvore usando a biblioteca squarify
squarify.plot(sizes=sizes, label=labels, alpha=.8)


```
